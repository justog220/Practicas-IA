## Visualizar los datos Geográficamente

Realice un scatterplot usando seaborn (Ver [documentación](https://seaborn.pydata.org/generated/seaborn.scatterplot.html)) para visualizar los datos en función de la longitud (eje x) y latitud (eje y).

```python
sns.scatterplot(data=housing, x='longitude', y='latitude')
```

Modifique el gráfico para modificar el radio de los círculos en función de la población en cada distrito y el color de los círculos represente el precio.

```python
sns.scatterplot(data=housing, x='longitude', y='latitude', hue='median_house_value', size='population', sizes=(20, 200))
```

¿A qué conclusión puede llegar del gráfico respecto a cómo varían los precios en función de la ubicación y en función de la población?

Como conclusión, interpretamos a partir del gráfico que más cerca del mar mayor será el precio y que la mayor parte de la población se concentra tambien cerca de él. Tanto al sur como al norte hay zonas donde se concentra gran parte de la población-.

## Datos faltantes

Una estrategia para trabajar con datos faltantes es imputar un valor (cero, la media o mediana del atributo)
A continuación, grafique el histograma del atributo con datos faltantes y decida que valor utilizar (media o mediana) en función de la distribución que tienen los datos.
Utilice la función **fillna** de pandas ([Documentación](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.fillna.html))

```python
sns.histplot(data=housing, x="total_bedrooms")
```

```python
housing["total_bedrooms"].fillna(value=housing["total_bedrooms"].mean(), inplace=True)
```

```python
housing.info()
```

```python
sns.histplot(data=housing, x="total_bedrooms")
```
