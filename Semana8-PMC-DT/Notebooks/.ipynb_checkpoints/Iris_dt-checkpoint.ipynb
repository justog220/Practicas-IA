```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from Notebooks.utiles import plot_decision_boundary
```

## Árboles de decisión

Los _árboles de decisión_ son algoritmos de Machine learning capaces de realizar tareas de clasificación y regresión. El algoritmo **CART** (_Classification and Regression Tree_) es el implementado por scikit-learn para el entrenamiento de los árboles de decisión. Esto produce sólo _árboles binarios_

```python
iris = load_iris()
X = iris.data[:,2:] #largo y ancho de pétalo
y = iris.target
X.shape
```

```python
print(iris.target_names)
print(iris.feature_names)
```

```python
tree_clf = DecisionTreeClassifier(max_depth=2) #podemos limitar diversos aspectos del arbol
tree_clf.fit(X,y)
```

```python
from sklearn import tree

tree.plot_tree(tree_clf,
               filled=True,
               rounded=True,
               class_names=['setosa', 'versicolor', 'virginica'],
               feature_names=iris.feature_names[2:])
plt.show()
```

Una de las muchas cualidades de los árboles de decisión es que requieren muy poca preparación de los datos. De hecho, no requieren escalado de los datos.

```python
plt.figure(figsize=(10, 6))
plot_decision_boundary(tree_clf, X, y, legend=True)
plt.plot([2.45, 2.45], [0, 3], "k-", linewidth=2)
plt.plot([2.45, 7.5], [1.75, 1.75], "k--", linewidth=2)
plt.plot([4.95, 4.95], [0, 1.75], "k:", linewidth=2)
plt.plot([4.85, 4.85], [1.75, 3], "k:", linewidth=2)
plt.text(1.40, 1.0, "Depth=0", fontsize=15)
plt.text(3.2, 1.80, "Depth=1", fontsize=13)
plt.text(4.05, 0.5, "(Depth=2)", fontsize=11)

plt.show()
```

```python
tree_clf = DecisionTreeClassifier(max_depth=3)
tree_clf.fit(X,y)

tree.plot_tree(tree_clf,
               filled=True,
               rounded=True,
               class_names=['setosa', 'versicolor', 'virginica'],
               feature_names=iris.feature_names[2:])
plt.show()
```
