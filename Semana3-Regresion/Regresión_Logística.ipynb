### Regresión Logística:

A pesar de su nombre, es un algoritmo de clasificación y no regresión.

El modelo de regresión logística se utiliza para clasificación binaria pero puede generalizarse a la clasificación multiclase, conocida como regresión logística multinomial o regresión softmax.

En este caso, la función que vamos a utilizar para la representación del modelo es la _función sigmoidea_ a la cual nos referimos como función de activación: 

$$\sigma(z) = \frac{1}{1+ e^{-z}}$$

Recordemos que $z=z(x, w)$ en el caso de dependencia con una variable, es la entrada o combinación lineal de los pesos.

### Función sigmoidea

```python
import matplotlib.pyplot as plt
import numpy as np

def sigmoid(z):
    return 1.0 / (1.0 + np.exp(-z))

z = np.arange(-8, 8, 0.1)
#print(len(z))
sigma_z = sigmoid(z)

plt.figure(figsize=(7,4))
plt.plot(z, sigma_z)
# Eje x
plt.xlim(-8,8)
plt.axvline(0.0, color='k')
plt.xlabel('z')
plt.ylabel('$\sigma (z)$')

# Eje y
plt.ylim(-0.1, 1.1)
plt.yticks([0.0, 0.5, 1.0])
ax = plt.gca()
ax.yaxis.grid(True)

plt.tight_layout()
plt.show()
```

Podemos ver que la función toma valores entre 0 y 1, además: $$\sigma(z) < 0.5 \qquad si \quad z<0$$ $$\sigma(z) > 0.5 \qquad si \quad z>0 $$

En la regresión logística usamos este "umbral" para predecir la probabilidad que una muestra pertenezca a una clase u otra. Nuestra predicción se puede representar:

$$\hat{y} = \begin{cases} y=1 & \text{si} & \sigma(z) \geq 0.5 \\ y=0 & \text{si} & \sigma(z) < 0.5 \end{cases}$$

La salida de la función sigmoidea puede interpretarse como la probabilidad de que una muestra particular pertenezca a una clase. Posteriormente, mediante una función umbral obtenemos una salida categórica.

Existen aplicaciones donde es necesario predecir tanto la clase de salida como la probabilidad de esa salida (ej. predicción del clima). En estas aplicaciones, la regresión logística resulta particularmente útil.

### Función de costo

La función de costo utilizada en regresión logística es la siguiente:

$$ J(w) = \frac{1}{n}\sum\limits_{i=1}^{n}[-y^{(i)}log{(\sigma{(z^{(i)})})} - (1-y^{(i)})log{(1-\sigma{(z^{(i)})})}]$$

No utilizamos la suma de los errores cuadráticos como función de costo porque tendríamos como resultado una función no convexa, esto haría que el algoritmo del descenso del gradiente no converja.

Si calculamos la derivada de la función de costo, obtendremos casi la misma regla de actualización de pesos que en la regresión lineal:

$${w}:= {w} + \eta(\nabla_w{J})$$

$$\nabla_w{J} = \sum\limits_{i=1}^{n}(y^{[i]} - \hat{y}^{[i]})x^{[i]}$$

Sin embargo, hay que tener en cuenta que la representación del modelo es distinta al igual que la función de costo.

### Ejemplo Dataset Iris

```python
import pandas as pd
#Cargo el Dataset
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', header=None)
df
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from matplotlib.colors import ListedColormap
```

```python
def plot_decision_regions(X, y, clasificador, test_idx=None, resolution=0.02):
    
    # setup marker generator and color map
    markers = ('s', 'o', '^', 'v', 'x')
    colors = ('red', 'blue', 'lightgreen', 'gray', 'cyan')
    cmap = ListedColormap(colors[:len(np.unique(y))])

    # plot the decision surface
    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1

    
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
                           np.arange(x2_min, x2_max, resolution))
    
    Z = clasificador.predict(np.array([xx1.ravel(), xx2.ravel()]).T)

    Z = Z.reshape(xx1.shape)

    plt.contourf(xx1, xx2, Z, alpha=0.3, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())

    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y == cl, 0], 
                    y=X[y == cl, 1],
                    alpha=0.8, 
                    c=colors[idx],
                    marker=markers[idx], 
                    label=cl, 
                    edgecolor='black')

    # highlight test samples
    if test_idx:
        # plot all samples
        X_test, y_test = X[test_idx, :], y[test_idx]

        plt.scatter(X_test[:, 0],
                    X_test[:, 1],
                    c="None",
                    edgecolor='black',
                    alpha=1,
                    linewidth=1,
                    marker='o',
                    s=100, 
                    label='test set')
```

```python
y = df.iloc[0:100, 4].values
#convertimos las etiquetas de clases en  0 (Iris-setosa)  y 1 (Iris-versicolor)
y = np.where(y == 'Iris-setosa', 0, 1)
# columna 0 : longitud de sépalo, columna 2 : longitud de pétalo
X = df.iloc[0:100, [0, 2]].values
print(X.shape)
print(y.shape)
```

```python
lrgd = LogisticRegression(max_iter=1000, random_state=1)
_ = lrgd.fit(X , y)
```

```python
y
```

```python
plt.figure(figsize=(7,5))
plot_decision_regions(X,y, lrgd)
plt.xlabel('longitud de sépalo')
plt.ylabel('longitud de pétalo')
plt.xlim([4, 7.2])
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```

### Usando 3 clases

```python
from sklearn.preprocessing import LabelEncoder
# Utilizo las 3 clases y cambio variables categóricas a numéricas
le = LabelEncoder()
y = df.iloc[:, 4].values
y = le.fit_transform(y)
```

```python
X = df.iloc[:, [2, 3]].values
print(X.shape) 
print(y.shape) 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1, stratify=y)
sc = StandardScaler()
sc.fit(X_train)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
X_train.shape
```

```python
X_combined_std = np.vstack((X_train_std, X_test_std))
y_combined = np.hstack((y_train, y_test))
lr = LogisticRegression(C=100.0, random_state=1, solver='lbfgs', multi_class='ovr')
lr.fit(X_train_std, y_train)
plt.figure(figsize=(9,7))
plot_decision_regions(X_combined_std, y_combined, clasificador=lr, test_idx=range(105, 150))
plt.xlabel('longitud de sépalo')
plt.ylabel('longitud de pétalo')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```
