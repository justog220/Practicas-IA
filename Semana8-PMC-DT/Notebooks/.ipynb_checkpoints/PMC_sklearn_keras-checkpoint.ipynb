```python
import numpy as np
import scipy.io as sio
import matplotlib.pyplot as plt
import sys
```

## Usando MLPClassifier de sklearn

```python
from sklearn.neural_network import MLPClassifier

datos_mat = sio.loadmat('../data/datos_xor.mat')

X = datos_mat['X']
y = datos_mat['Y']
y = np.where(y==0.1, 0, 1)
y = y.flatten() 

pmc2 = MLPClassifier(hidden_layer_sizes=(2,), activation='logistic', learning_rate_init=0.2, solver='adam', max_iter=500, random_state=37)
pmc2.fit(X,y)

#hiden_layer_sizes=(2, ) es una sola capa oculta, con dos nodos
#solver='adams' significa el metodo de reoslucion que se utiliza, en este caso es de adams, nosotros vimos descenso del emc
_ = plt.plot(pmc2.loss_curve_)
```

```python
w_h = np.hstack((pmc2.coefs_[0], pmc2.intercepts_[0][:,np.newaxis]))
w_o = np.vstack((pmc2.coefs_[1],pmc2.intercepts_[1])).T
print(w_h)
print(w_o)
```

```python
m1 = - w_h[0][0]/w_h[0][1]
m2 = - w_h[1][0]/w_h[1][1]

b1 = - w_h[0][2]/w_h[0][1]
b2 = - w_h[1][2]/w_h[1][1]

y = y[:, np.newaxis]
# Graficamos las 2 rectas junto a los datos
#------------------------------------------------------------------------------------
plt.figure()
plt.scatter(X[ y[:,0]==0 , 0], X[ y[:,0]==0 , 1], c = 'skyblue', label='clase 0', marker='*', s=50)
plt.scatter(X[y[:,0]==1 , 0], X[y[:,0]==1 , 1], c= 'salmon', label='clase 1', marker='o', s=50)

x2 =np.linspace(0,1,100)
rect1= m1*x2 + b1
plt.plot(x2, rect1)

rect2= m2*x2 + b2
plt.plot(x2, rect2)
plt.show()
```

## Usando Keras

Biblioteca para implementar redes neuronales, diseada especialmente para aprendizaje profundo. Se ejecuta sobre TensorFlow, una biblioteca de aprendizaje automtico basado en redes neuronales de aprendizaje profundo donde las operaciones en las redes se realizan sobre arrays multidimensionales o tensores.

```python
X = datos_mat['X']
y = datos_mat['Y']
y = np.where(y==0.1, 0, 1)
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
X.shape
```

```python
keras.utils.set_random_seed(0)

# model = keras.models.Sequential()
# model.add( keras.layers.Flatten(input_shape=[2]))
# model.add( keras.layers.Dense(2, activation='sigmoid', name='capa_oculta') )
# model.add( keras.layers.Dense(1, activation='sigmoid', name='capa_salida') )

model = keras.models.Sequential(
    [
        keras.layers.Flatten(input_shape=[2]),
        keras.layers.Dense(2, activation='sigmoid', name='capa_oculta'),
        keras.layers.Dense(1, activation='sigmoid', name='capa_salida')
    ]
)

pesos, bias = model.get_layer('capa_oculta').get_weights()
print("Pesos: ", pesos)
print("Bias: ", bias)
```

```python
model.summary()
```

```python
opt = keras.optimizers.Adam(learning_rate=0.1) #Fijamos cual va a ser la funcion de optimizacion
loss = keras.losses.BinaryCrossentropy() #Cual es el error que quiero minimizar, BinnaryCrossentropy porque es un problema binario
model.compile(loss=loss, optimizer=opt, metrics=['accuracy'])
```

```python
y=y.T
history = model.fit(X, y, epochs=50, validation_data=(X, y) )
#podemos pasarle datos de vlaidacion que sean distintos, en este caso como son pocos les pasamos los mismos.
```

```python
w_h, b_h = model.get_layer('capa_oculta').get_weights()
w_h = np.vstack((w_h,b_h)).T
w_h
```

```python
m1 = - w_h[0][0]/w_h[0][1]
m2 = - w_h[1][0]/w_h[1][1]

b1 = - w_h[0][2]/w_h[0][1]
b2 = - w_h[1][2]/w_h[1][1]

# Graficamos las 2 rectas junto a los datos
#------------------------------------------------------------------------------------
plt.figure()
plt.scatter(X[ y[:,0]==0 , 0], X[ y[:,0]==0 , 1], c = 'skyblue', label='clase 0', marker='*', s=50)
plt.scatter(X[y[:,0]==1 , 0], X[y[:,0]==1 , 1], c= 'salmon', label='clase 1', marker='o', s=50)

x2 =np.linspace(0,1,100)
rect1= m1*x2 + b1
plt.plot(x2, rect1)

rect2= m2*x2 + b2
plt.plot(x2, rect2)
plt.show()
```
