```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## 1.- Cree y grafique una distribución uniforme y una distribución gaussiana univariadas usando subplots e histogramas.

```python
distUnif = np.random.rand(200)
distUnif
```

```python
distGauss = np.random.randn(200)
distGauss[[]]
```

```python
plt.subplot(2, 2, 1)
plt.plot(distUnif)
plt.title("Distribución uniforme")
plt.grid(True)
plt.subplot(2, 2, 2)
plt.plot(distGauss)
plt.title("Distribución gaussiana")
plt.grid(True)
plt.subplot(2, 2, 3)
#plt.title("Hist. dist uniforme")
plt.xlabel("Valor")
plt.ylabel("Frecuencia")
plt.hist(distUnif, bins = 10, rwidth=0.8)
plt.subplot(2, 2, 4)
#plt.title("Hist. dist gauss")
plt.xlabel("Valor")
plt.ylabel("Frecuencia")
plt.hist(distGauss, bins = 10, rwidth=0.8)
plt.show
```

2.- Cargue los datos del archivo "datos.csv" en la carpeta data, analice los valores de la columna dataset para separar los datos en 4 grupos y grafíquelos usando scatterplot y subplots.

```python
data = pd.read_csv("../data/datos.csv")
data
```

```python
datasetI = data.query("dataset == 'I'") # data[data["dataset"] == 'I'], podríamos haberlo graficado de una no crear otro dataframe
datasetI
```

```python
datasetII = data.query("dataset == 'II'")
datasetII
```

```python
datasetIII = data.query("dataset == 'III'")
datasetIII
```

```python
datasetIV = data.query("dataset == 'IV'")
datasetIV
```

```python
plt.subplot(221)
plt.title("Dataset I")
plt.scatter(datasetI["x"], datasetI["y"])
plt.subplot(222)
plt.title("Dataset II")
plt.scatter(datasetII["x"], datasetII["y"])
plt.subplot(223)
plt.title("Dataset III")
plt.scatter(datasetIII["x"], datasetIII["y"])
plt.subplot(224)
plt.title("Dataset IV")
plt.scatter(datasetIV["x"], datasetIV["y"])
plt.show()
```

## 3.- Grafique la siguiente recta junto con el texto como se muestra en la figura y muestre por consola su pendiente.


<img src="https://drive.google.com/uc?export=view&id=1HLm6NpD4lenot5LR2BXI4SUDJbjCrZXR" width = 450>

```python
a = (1, 1)
b = (7, 4)
m = (b[1] - a[1]) / (b[0] - a[0])
m
```

```python
x = np.linspace(-1, 9, 500)
y = m * x + 0.5
plt.plot(x, y)
plt.title("Slope = rise / run")
plt.xlim(0, 8.2)
plt.ylim(0, 5.2)
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.plot(a[0], a[1], "bo")
plt.text(a[0] - 0.2, a[1], "A", ha="right")
plt.plot(b[0], b[1], "bo")
plt.text(b[0] - 0.2, b[1], "B", ha="right")
x = np.linspace(1, 7, 500)
y = x/x
plt.plot(x, y, "k--")
y = np.linspace(1, 4, 500)
x = np.full(500, 7)
plt.plot(x, y, "k--")
plt.text(4, 1-0.2, "run", ha="center")
plt.text(7+0.8, 2.5, "rise", ha="right")
plt.show
```
